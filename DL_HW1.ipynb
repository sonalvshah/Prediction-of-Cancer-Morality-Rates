```python

#outlier analysis 
#indentifying the outliers 
import scipy.stats as stats

# Calculate Z-scores
df_main['z_score'] = (df_main['TARGET_deathRate'] - df_main['TARGET_deathRate'].mean()) / df_main['TARGET_deathRate'].std()

# Calculate quantiles for outlier thresholds
q_high = df_main['TARGET_deathRate'].quantile(0.95)
q_low = df_main['TARGET_deathRate'].quantile(0.05)

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Outlier Analysis of Cancer Death Rate', fontsize=16)

# Box Plot
sns.boxplot(x=df_main['TARGET_deathRate'], color='lightcoral', ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Cancer Death Rate')
axes[0, 0].set_xlabel('Cancer Death Rate')

# Scatter Plot with Z-scores
scatter = axes[0, 1].scatter(df_main.index, df_main['TARGET_deathRate'], c=np.abs(df_main['z_score']) > 3, cmap='coolwarm', edgecolor='k')
axes[0, 1].set_title('Scatter Plot of Cancer Death Rate with Z-scores')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Cancer Death Rate')
fig.colorbar(scatter, ax=axes[0, 1], label='Outlier Indicator')

# Violin Plot
sns.violinplot(x=df_main['TARGET_deathRate'], color='lightcoral', ax=axes[0, 2])
axes[0, 2].set_title('Violin Plot of Cancer Death Rate')
axes[0, 2].set_xlabel('Cancer Death Rate')

# QQ Plot
stats.probplot(df_main['TARGET_deathRate'], dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('QQ Plot of Cancer Death Rate')

# Rug Plot
sns.rugplot(df_main['TARGET_deathRate'], color='lightcoral', ax=axes[1, 1])
axes[1, 1].set_title('Rug Plot of Cancer Death Rate')
axes[1, 1].set_xlabel('Cancer Death Rate')

# Hide the empty subplot (bottom-right corner)
axes[1, 2].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import scipy.stats as stats

# Calculate Z-scores
df_main['z_score'] = (df_main['TARGET_deathRate'] - df_main['TARGET_deathRate'].mean()) / df_main['TARGET_deathRate'].std()

# Define a Z-score threshold for identifying outliers
threshold = 3

# Filter out outliers
df_no_outliers = df_main[np.abs(df_main['z_score']) <= threshold]

# Plotting before and after outlier removal
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Comparison of Cancer Death Rate Before and After Outlier Removal', fontsize=16)

# Before removing outliers
sns.boxplot(x=df_main['TARGET_deathRate'], color='lightcoral', ax=axes[0, 0])
axes[0, 0].set_title('Box Plot Before Outlier Removal')
axes[0, 0].set_xlabel('Cancer Death Rate')

sns.scatterplot(x=df_main.index, y=df_main['TARGET_deathRate'], ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Scatter Plot Before Outlier Removal')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Cancer Death Rate')

sns.violinplot(x=df_main['TARGET_deathRate'], color='lightcoral', ax=axes[0, 2])
axes[0, 2].set_title('Violin Plot Before Outlier Removal')
axes[0, 2].set_xlabel('Cancer Death Rate')

# After removing outliers
sns.boxplot(x=df_no_outliers['TARGET_deathRate'], color='lightcoral', ax=axes[1, 0])
axes[1, 0].set_title('Box Plot After Outlier Removal')
axes[1, 0].set_xlabel('Cancer Death Rate')

sns.scatterplot(x=df_no_outliers.index, y=df_no_outliers['TARGET_deathRate'], ax=axes[1, 1], color='skyblue')
axes[1, 1].set_title('Scatter Plot After Outlier Removal')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('Cancer Death Rate')

sns.violinplot(x=df_no_outliers['TARGET_deathRate'], color='lightcoral', ax=axes[1, 2])
axes[1, 2].set_title('Violin Plot After Outlier Removal')
axes[1, 2].set_xlabel('Cancer Death Rate')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
```

```python
from datetime import datetime
```

```python
x=df_main.drop(['TARGET_deathRate'],axis=1)
y=df_main['TARGET_deathRate']

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.20,random_state=0)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

lreg=LinearRegression()

lreg.fit(x_train,y_train)
y_pred = lreg.predict(x_test)

print(datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
